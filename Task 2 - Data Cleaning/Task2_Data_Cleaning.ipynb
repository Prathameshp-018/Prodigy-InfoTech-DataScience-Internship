# Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Step 2: Load Dataset

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

# Step 3: Check Dataset Info

```python
df.info()
df.describe()
```

# Step 4: Check for Missing Values

```python
df.isnull().sum()
```

# Step 5: Check for Duplicates

```python
df.duplicated().sum()
```

# Step 6: Fix Inconsistent Data

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df['gender'] = df['gender'].str.lower()
```

# Step 7: Handle Missing Values

```python
df = df.fillna(df.mean(numeric_only=True))
```

# Step 8: Outlier Detection

```python
df.to_csv("StudentsPerformance_Cleaned.csv", index=False)

# Step 8: Outlier Detection
plt.figure(figsize=(8,4))
sns.boxplot(data=df[['math_score', 'reading_score', 'writing_score']])
plt.title('Outlier Detection in Scores')
plt.show()

```
