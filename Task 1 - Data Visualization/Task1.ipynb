# step 1: Load the Dataset

# Titanic dataset is built into Seaborn, no download needed

```python
df = sns.load_dataset('titanic')
```

# step 2: Clean & Prepare Data

# Remove rows with missing values for easier plotting

```python
df = df.dropna(subset=['age', 'sex', 'class', 'fare', 'survived'])
```

# Check for missing values again

```python
print("\n Missing Values After Cleaning:")
```

```python
print(df.isnull().sum())
```

# Step 3: Visualizations

# 1️ Gender Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='sex', data=df, palette='coolwarm')
plt.title('Gender Distribution of Titanic Passengers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# 2️ Survival by Gender

```python
plt.figure(figsize=(6,4))
sns.barplot(x='sex', y='survived', data=df, palette='viridis')
plt.title('Survival Rate by Gender')
plt.ylabel('Average Survival Rate')
plt.show()
```

# 3️ Age Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=20, kde=True, color='teal')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# 4️ Survival Rate by Passenger Class

```python
plt.figure(figsize=(6,4))
sns.barplot(x='class', y='survived', data=df, palette='magma')
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Average Survival Rate')
plt.show()
```

# 5️ Box Plot - Age vs Class

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='class', y='age', data=df, palette='Set2')
plt.title('Age Distribution Across Passenger Classes')
plt.show()
```

# 6️ Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Numeric Features')
plt.show()
```
